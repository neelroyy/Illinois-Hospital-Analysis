```python
import pandas as pd
```

```python
il_hospital_lat_long_df = pd.read_csv('data/il_hospital_coord_rev.csv')
# il_hospital_lat_long_df
il_hospital_lat_long_df1=il_hospital_lat_long_df[['hospital_name','latitude','longitude','google_stars']]
il_hospital_lat_long_df1
```

```python
hospital_data_clean = pd.read_csv('data/hospital_data_clean.csv')
# hospital_data_clean
hospital_data_df = hospital_data_clean.rename(columns = {'Hospital Name': 'hospital_name','City': 'city',
                                                 'TotalDischarges': 'totaldischarges','PatientDays': 'patientdays', 
                                                 'Gross PatientRevenue ($000)': 'gross patientrevenue ($000)'})
hospital_data_df
```

```python
hospital_data_final = pd.merge(hospital_data_df, il_hospital_lat_long_df1, on=['hospital_name'])
hospital_data_final
```

```python
hospital_data_final.to_csv(r'data/hospital_data_final.csv', index=False)
```

```python
bottom20 = hospital_data_final.sort_values(by='google_stars', ascending=True)
bottom20hospitals = bottom20['hospital_name']
bottom20hospitals.head(20)
```
