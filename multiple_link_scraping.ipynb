```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
# setup spliner
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.ahd.com/free_profile/140147/_Carle_Richland_Memorial_Hospital/Olney/Illinois/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
# Obtain information from tag <table>
profile = soup.find('table', class_='nomargin')
print(profile.prettify())
```

```python
details = profile.find('tr').text
# details
result = details.split("\n")
for i in range(len(result)):
    if(result[i]):
        print(result[i])
    #result[i] = result[i].replace(" ", "")
    #print(result[i])
# result_set = set(result)
# print(result)
```
