```python
# import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Create an URL object
url = 'https://www.ahd.com/states/hospital_IL.html'
r = requests.get(url)

#check status/response 
# print(r.status_code)

# soup = BeautifulSoup(r.text, 'html.parser')
# parser-lxml = Change html to Python friendly format
soup = BeautifulSoup(r.text, 'lxml')
print(soup.prettify())
```

```python
# Obtain information from tag <table>
hospital_table = soup.find('table')
hospital_table
```

```python
# Obtain every title of columns with tag <th>
headers = []
for i in hospital_table.find_all('th'):
 title = i.text
 headers.append(title)
headers
```

```python
# Create a dataframe
hospital_data = pd.DataFrame(columns = headers)
hospital_data
```

```python
# Create a for loop to fill mydata
for j in hospital_table.find_all('tr')[1:]:
 row_data = j.find_all('td')
 row = [i.text for i in row_data]
 length = len(hospital_data)
 hospital_data.loc[length] = row
hospital_data
```

```python
hospital_data['StaffedBeds'] = hospital_data['StaffedBeds'].str.replace(',', '')
hospital_data['TotalDischarges'] = hospital_data['TotalDischarges'].str.replace(',', '')
hospital_data['PatientDays'] = hospital_data['PatientDays'].str.replace(',', '')
hospital_data['Gross PatientRevenue ($000)'] = hospital_data['Gross PatientRevenue ($000)'].str.replace(',', '')
hospital_data['Gross PatientRevenue ($000)'] = hospital_data['Gross PatientRevenue ($000)'].str.replace('$', '')
hospital_data['StaffedBeds'] = hospital_data['StaffedBeds'].astype(int)
hospital_data['TotalDischarges'] = hospital_data['TotalDischarges'].astype(int)
hospital_data['PatientDays'] = hospital_data['PatientDays'].astype(int)
hospital_data['Gross PatientRevenue ($000)'] = hospital_data['Gross PatientRevenue ($000)'].astype(int)
hospital_data
```

```python
hospital_data_clean = hospital_data[hospital_data['StaffedBeds']>0]
hospital_data_clean
```
